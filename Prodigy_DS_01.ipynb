```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('2015.csv')
df.head(10)
```

```python
plt.figure(figsize=(12, 6))
df['Happiness Score Rounded'] = df['Happiness Score'].round()

# Distribution plot for Happiness Score across different regions
sns.boxplot(x='Happiness Score Rounded',y='Region', data = df)

plt.title('Distribution of Happiness Score Across Regions')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(16, 8))
df['Economy (GDP per Capita)'] = df['Economy (GDP per Capita)'].round(2)
sns.boxplot(y='Region', x='Economy (GDP per Capita)', data=df, palette='viridis', showfliers=False)
sns.violinplot(y='Region', x='Economy (GDP per Capita)', data=df, palette='Pastel1', inner='quartile', cut=0)

plt.show()
```
